```python
percentages = {"Rock":0, "Electronic": 0, "Pop": 0.4, "Jazz": 0.1, "RnB":0, "Rap":0.05, "Metal":0, "Country": 0, "Blues": 0, "Reggae":0, "Folk":0, "Punk":0, "Latin": 0.5, "World":0, "New Age":0}
```

```python
def getSongByPercentages(percentages):
    diffSum = 0
    track_id = 0
    for i, row in results.iterrows():
        localSum = 0
        for percentage in percentages:
            diff = row[percentage] - percentages[percentage]
            localSum = localSum + diff
    
        if (diffSum > localSum):
            diffSum = localSum
            track_id = row["track_id"]
            
    return track_id
```

```python
theSong = getSongByPercentages(percentages)
theSong
```
